```python
import os
from directories import Dicrectories
from tools import Tools

num_features = 0
num_clauses = 0
knowledge_directory = Dicrectories.knowledge
for file_name in os.listdir(knowledge_directory):
    if file_name.endswith(".pkl"):
        file_path = os.path.join(knowledge_directory , file_name)
        clauses = Tools.read_pickle_data(file_path)
        num_clauses = num_clauses + len(clauses)
        for clause in clauses:
            num_features = num_features + len(clause[1])
if num_clauses > 0:
    avg_features = num_features / num_clauses
    print(f"Average number of features: {avg_features:.2f}")
```
