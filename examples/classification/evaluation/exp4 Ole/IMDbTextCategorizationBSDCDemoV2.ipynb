```python
from contextlib import redirect_stdout
import HypervectorNLP as HypervectorNLP
```

```python
scaling = 2
p = 0.0005
hypervector_size = 10000*scaling
bits = int(p * hypervector_size)

clauses = 10000
T = 100
s = 20.0
#p = 0.25*1.0/np.sqrt(hypervector_size)
MAX_NGRAM = 2
NUM_WORDS=5000
INDEX_FROM=2
FEATURES=5000

scaling_factors_to_test = [0.5, 1.0, 1.5, 2.0, 2.5]
for scaling in scaling_factors_to_test:
    hypervector_size = 10000*scaling
    bits = int(p * hypervector_size)
	with open(HypervectorNLP.get_test_name("HV"), 'w') as file, redirect_stdout(file):
		encoding = HypervectorNLP.generate_encoding_hypervector(hypervector_size, bits, FEATURES)
		train_y, test_y, X_train, X_test = HypervectorNLP.get_processed_data(NUM_WORDS, INDEX_FROM)
		X_train_org, X_test_org = HypervectorNLP.apply_feature_selection(FEATURES, train_y, X_train, X_test)
		print("Processing dataset with scaling factor of:", scaling)
		X_train, X_test, Y_train, Y_test = HypervectorNLP.encode_input_text(hypervector_size, encoding, train_y, test_y, X_train_org, X_test_org)
		HypervectorNLP.apply_simulated_annealing(clauses, T, s, X_train, X_test, Y_train, Y_test)
```
